```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Iris.csv')
```

```python
#Find out number of tuples(rows) by checking dataframe
data.info()
```

```python
#Find out features' names and their data type(columns)
data.dtypes
```

```python
#finding out number of unique species under class feature
data["class"].nunique()
```

```python
#adding new row and calculate range for data.describe.
ds = data.describe()
ds.loc['range'] = ds.loc['max'] - ds.loc['min']
display(ds)
```

```python
#basic scatter plot for sepal length and sepal width
data.plot.scatter(x='sepal_length', y='sepal_width')
```

```python
#Creating histogram for each features. Bins are number of bars we wanted to show.
data.hist(bins=50, figsize=(20,15))
```

```python
#Using seaborn library to better visualize boxplots for each features
#sns.boxplot(data=data,color='green')
data.boxplot(column=['sepal_length', 'sepal_width','petal_length','petal_width'], by='class')
```

```python
#Making basic scatter plot matrix for the data
pd.plotting.scatter_matrix(data)
```
